```python
import numpy as np # utilizaremos el metodo np.random.choice() para la asignacion random de valores
```

```python
df['Sex '].unique()
```

```python
def unificar_sexo (sex):
    lista = ['F','M']
    if sex == 'F':
        return 'F'
    elif sex == 'M':
        return 'M'
    elif sex == 'lli':
        np.random.choice(lista)
    else:
        nan
```

```python
df['Sex (M/F)'] = df['Sex '].apply(unificar_sexo)
```
